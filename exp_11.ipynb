```python
import cv2
import matplotlib.pyplot as plt

# Define the image path
image_path = '/content/images.jpg'

# Load the image
original_image = cv2.imread(image_path)

# Check if the image was loaded successfully
if original_image is None:
    print(f"Error: Could not load image from {image_path}")
else:
    # Perform a horizontal flip (rotation around y-axis by 180 degrees)
    # cv2.flip(src, flipCode)
    # flipCode = 0 for flipping around the x-axis (vertical flip)
    # flipCode > 0 for flipping around the y-axis (horizontal flip)
    # flipCode < 0 for flipping around both axes
    rotated_image_y_axis = cv2.flip(original_image, 1)

    # Convert images from BGR to RGB for displaying with Matplotlib
    original_image_rgb = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)
    rotated_image_y_axis_rgb = cv2.cvtColor(rotated_image_y_axis, cv2.COLOR_BGR2RGB)

    # Display the original and rotated images
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.imshow(original_image_rgb)
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(rotated_image_y_axis_rgb)
    plt.title('Rotated 180-degree along Y-axis (Horizontal Flip)')
    plt.axis('off')

    plt.show()
```
