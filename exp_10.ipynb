```python
from PIL import Image

# Define the path to the image
img_path = '/content/images.jpeg'

# Load the image
original_img = Image.open(img_path)

# Perform a 90-degree clockwise rotation
# In PIL, positive angles are counter-clockwise, so -90 degrees is clockwise.
rotated_img = original_img.rotate(-90, expand=True)

# Display the original and rotated images
print("Original image:")
display(original_img)

print("Rotated image (90 degrees clockwise):")
display(rotated_img)
```
