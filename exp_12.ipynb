```python
from PIL import Image
import matplotlib.pyplot as plt

# Define the image path
image_path = '/content/images.jpg'

# Load the image
original_image = Image.open(image_path)

# Perform a 270-degree clockwise rotation.
# In Pillow, a 270-degree clockwise rotation is equivalent to a 90-degree counter-clockwise rotation.
# The `Image.ROTATE_90` method performs a 90-degree counter-clockwise rotation.
rotated_image = original_image.transpose(Image.ROTATE_90)

# Display both original and rotated images
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(original_image)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(rotated_image)
plt.title('Rotated 270° Clockwise')
plt.axis('off')

plt.tight_layout()
plt.show()
```
